```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import re
from sklearn.preprocessing import LabelEncoder
```

```python
wines = pd.read_csv('winemag-data-130k-v2.csv')
```

```python
wines.head()
```

```python
wines.isnull().values.any()
```

```python
def nullscan(df):
    nulls = []
    columns = df.columns 
    for c in columns:
        nulls.append(df[c].isnull().sum()/df.shape[0])

    fig = plt.figure(figsize=(10,8))
    ax = sns.barplot(x=columns, y=nulls)
    ax.set_xticklabels(columns, rotation=90)
    ax.set_ylim([-0.05, 1])
    plt.show()
```

```python
nullscan(wines)
```

```python
wines['price'] = wines['price'].fillna(wines['price'].mean())
wines = wines.fillna({'designation': 'NAN', 'province': 'NAN', 'region_1': 'NAN', 'region_2': 'NAN', 'taster_name': 'NAN'})
wines = wines.drop(['Unnamed: 0', 'taster_twitter_handle'], axis=1)
```

```python
nullscan(wines)
```

```python
wines['year'] = wines['title'].apply(lambda x: re.search('19\d{2}|20\d{2}', x).group() if re.search('19\d{2}|20\d{2}', x) else np.NaN)
wines['year'] = wines['year'].astype(float)
```

```python
wines.select_dtypes('object').columns
```

```python
le = {}
for c in wines.select_dtypes('object').columns:
    le[c] = LabelEncoder()
    le[c].fit(wines[c])
    wines[c] = le[c].transform(wines[c])
```
